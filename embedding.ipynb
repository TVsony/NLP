```python
from tensorflow.keras.preprocessing.text import one_hot
from tensorflow.keras.utils import pad_sequences
from tensorflow.keras.layers import Embedding
from tensorflow.keras.models import Sequential
import numpy as np
```

```python
sent=[ 'the glass of milk',
      'the glass of juice',
      'the cup of tea',
      'I am a good boy',
      'I am a good developer',
      'understand the meaning of words',
      'your video are good',
      ]
```

```python
sent
```

```python
# Define vocabulary size
voc_size = 10000
```

```python
# One Hot Representation
one_hot_repr = [one_hot(words, voc_size) for words in sent]
one_hot_repr # given the index of the vectors 
```

```python
# Word emdedding repsentation 
from tensorflow.keras.layers import Embedding
# from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.utils import pad_sequences
from tensorflow.keras.models import Sequential
```

```python
import numpy as np
```

```python
# We need to make all sentence are equal sizes other wise we not able to train model 
# Make all sentences equal sizes using padding
sent_length = 8  # Desired length of all sequences
embedded_docs = pad_sequences(one_hot_repr, padding='pre', maxlen=sent_length)
print(embedded_docs)
```

```python
# feature representation 
dim=10
```

```python
# Build the model
model = Sequential()

# Add embedding layer
model.add(Embedding(voc_size, dim, input_length=sent_length))

# Compile the model
model.compile(optimizer='adam', loss='mse')
```
